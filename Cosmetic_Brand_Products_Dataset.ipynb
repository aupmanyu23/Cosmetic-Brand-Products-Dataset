```python
columns_to_drop = ['image_link', 'product_link', 'website_link', 'product_api_url', 'api_featured_image']
df_cleaned = df.drop(columns=columns_to_drop)

df_cleaned.head()
```

```python
# Check unique values in price_sign and currency
print(df_cleaned['price_sign'].unique())
print(df_cleaned['currency'].unique())
```

```python
# Check the value counts for price_sign and currency
print("Price Sign Value Counts:")
print(df_cleaned['price_sign'].value_counts())
print("\nCurrency Value Counts:")
print(df_cleaned['currency'].value_counts())
```

```python
# Drop price_sign and keep currency
df_cleaned.drop(columns=['price_sign'], inplace=True)

df_cleaned.isnull().sum()
```

```python
# Impute categorical columns with 'Unknown'
df_cleaned['brand'].fillna('Unknown', inplace=True)
df_cleaned['category'].fillna('Unknown', inplace=True)
df_cleaned['description'].fillna('No description', inplace=True)

# Impute numerical columns with mean/median
df_cleaned['price'].fillna(df_cleaned['price'].mean(), inplace=True)
df_cleaned['rating'].fillna(df_cleaned['rating'].median(), inplace=True)

# Impute missing currency values with 'Unknown'
df_cleaned['currency'].fillna('Unknown', inplace=True)

df_cleaned.isnull().sum()
```

```python
# detect and remove outliers
def detect_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5*IQR
  upper_bound = Q3 + 1.5*IQR

  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  print(f"Number of outliers in {column}: {outliers.shape[0]}")

  df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

  return df_filtered


df_cleaned = detect_outliers_iqr(df_cleaned, 'price')
df_cleaned = detect_outliers_iqr(df_cleaned, 'rating')

df_cleaned.shape
```

```python
df_cleaned.info()
```

```python
df_cleaned['created_at'] = pd.to_datetime(df_cleaned['created_at'])
df_cleaned['updated_at'] = pd.to_datetime(df_cleaned['updated_at'])
```

```python
df_cleaned['created_year'] = df_cleaned['created_at'].dt.year
df_cleaned['created_month'] = df_cleaned['created_at'].dt.month
df_cleaned['time_difference'] = (df_cleaned['updated_at'] - df_cleaned['created_at']).dt.total_seconds()

df_cleaned.drop(columns=['created_at','updated_at'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

# Initialize LabelEncoder and MinMaxScaler
le = LabelEncoder()
scaler = MinMaxScaler()

# Categorical columns to encode
categorical_columns = ['brand', 'name', 'currency', 'description', 'category', 'product_type', 'tag_list', 'product_colors']

# Non-categorical columns to scale
non_categorical_columns = ['price', 'created_year', 'created_month', 'time_difference']

# Create a dictionary to store encoders
label_encoders = {}

# Apply label encoding to categorical columns
for col in categorical_columns:
    le = LabelEncoder()
    df_cleaned[col] = le.fit_transform(df_cleaned[col].astype(str))
    label_encoders[col] = le  # Save encoder for consistency

# Apply Min-Max scaling to non-categorical columns
df_cleaned[non_categorical_columns] = scaler.fit_transform(df_cleaned[non_categorical_columns])

# Check the result
df_cleaned.head()
```

**Brand Analysis**

**Pricing Insights**

```python
# Basic statistics for price
price_stats = df['price'].describe()
print(price_stats)

# Visualize the distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True, color='purple')
plt.title('Distribution of Prices Across Cosmetic Products')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Boxplot to show price distribution by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='price', data=df)
plt.title('Price Distribution by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 most expensive products
top_expensive = df[['name', 'brand', 'price']].sort_values(by='price', ascending=False).head(10)
print(top_expensive)
```

```python
# Analyze the most common product categories
category_counts = df['category'].value_counts()

# Display top categories
print(category_counts)

# Visualize the most common product categories
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.values, y=category_counts.index, palette='viridis')
plt.title('Most Common Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

**Tag analysis**

```python
from collections import Counter

# Flatten the 'tag_list' column, assuming it's stored as a list of tags
all_tags = df['tag_list'].apply(lambda x: x.strip('[]').replace("'", "").split(', ')).tolist()

# Count the frequency of each tag
tag_counts = Counter([tag for sublist in all_tags for tag in sublist])

# Convert to a DataFrame for easy visualization
tag_df = pd.DataFrame(tag_counts.items(), columns=['Tag', 'Count']).sort_values(by='Count', ascending=False)

# Display top tags
print(tag_df.head(10))

# Visualize the top tags
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Tag', data=tag_df.head(10), palette='coolwarm')
plt.title('Top 10 Most Common Tags')
plt.xlabel('Number of Products')
plt.ylabel('Tag')
plt.tight_layout()
plt.show()
```

```python
# Count the number of tags per product
df['num_tags'] = df['tag_list'].apply(lambda x: len(x.strip('[]').replace("'", "").split(', ')))

# Visualize the distribution of the number of tags per product
plt.figure(figsize=(10, 6))
sns.histplot(df['num_tags'], bins=10, kde=False, color='green')
plt.title('Distribution of the Number of Tags per Product')
plt.xlabel('Number of Tags')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

```python
# Analyze common tag combinations
tag_combinations = df['tag_list'].value_counts().head(10)
print(tag_combinations)
```

```python
# Explore tag frequency by brand
brand_tag_counts = df.groupby('brand')['tag_list'].apply(lambda x: ', '.join(x).strip('[]').replace("'", "").split(', '))
brand_tag_counts = pd.Series([tag for sublist in brand_tag_counts for tag in sublist])

# Display most common tags by brand
brand_tag_counts = brand_tag_counts.value_counts().head(10)
print(brand_tag_counts)
```

```python
#Temporal analysis
# Plot number of products created by year
plt.figure(figsize=(10, 6))
sns.countplot(x='created_year', data=df, palette='viridis')
plt.title('Number of Products Created per Year')
plt.xlabel('Year')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

```python
# Plot distribution of time differences (in seconds)
plt.figure(figsize=(10, 6))
sns.histplot(df['time_difference'], bins=20, kde=True, color='purple')
plt.title('Distribution of Time Difference Between Product Creation and Update')
plt.xlabel('Time Difference (seconds)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot number of products created by month
plt.figure(figsize=(10, 6))
sns.countplot(x='created_month', data=df, palette='coolwarm')
plt.title('Number of Products Created per Month')
plt.xlabel('Month')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Flatten the 'product_colors' column (assuming it's stored as a list of dictionaries with hex values)
all_colors = df['product_colors'].apply(lambda x: eval(x) if x != '[]' else []).tolist()

# Extract color names from the list of dictionaries
color_names = [color['colour_name'] for sublist in all_colors for color in sublist]

# Count the frequency of each color
color_counts = Counter(color_names)

# Convert to DataFrame for easy visualization
color_df = pd.DataFrame(color_counts.items(), columns=['Color', 'Count']).sort_values(by='Count', ascending=False)

# Display top colors
print(color_df.head(10))

# Visualize the most common colors
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Color', data=color_df.head(10), palette='coolwarm')
plt.title('Top 10 Most Common Colors in Cosmetic Products')
plt.xlabel('Number of Products')
plt.ylabel('Color')
plt.tight_layout()
plt.show()
```
